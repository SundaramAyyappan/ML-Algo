```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
```

```python
movie = pd.read_csv("E:/Userfiles/Desktop/Python/New folder/my_movies.csv")
movie
```

```python
movie1  =movie.iloc[:, 5:]
```

```python
frequent_itemsets_ap = apriori(movie1, min_support=0.1)
frequent_itemsets_ap
```

```python
frequent_itemsets_ap = apriori(movie1, min_support=0.1, use_colnames= True, verbose=1)
frequent_itemsets_ap
```

```python
rules_ap = association_rules(frequent_itemsets_ap, metric= "confidence", min_threshold=0.4)
rules_ap
```

```python
rules_ap[(rules_ap.support > 0.7) & (rules_ap.confidence > 0.7)].sort_values("confidence", ascending = False).shape
```

```python
rules_ap['lhs items'] = rules_ap['antecedents'].apply(lambda x:len(x))
rules_ap
rules_ap[rules_ap['lhs items']>2].sort_values('lift', ascending=False)
```

```python
rules_ap['antecedents_'] = rules_ap['antecedents'].apply(lambda a: ','.join(list(a)))
rules_ap['consequents_'] = rules_ap['consequents'].apply(lambda a: ','.join(list(a)))

pivot = rules_ap[rules_ap['lhs items']>2].pivot(index = 'antecedents_', 
                    columns = 'consequents_', values= 'confidence')
pivot
```

```python
sns.heatmap(pivot, annot = True)
plt.title('Heat Map - For Confidence Metric')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```
